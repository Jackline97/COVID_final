```python
import datetime
def transfer_start_expire_toperiod(create_date,expire_date):
    temp1 = []
    temp2 = []
    for i in create_date:
        if i!='0':
            date_time_obj = datetime.datetime.strptime(i[2:-2],'%Y-%m-%d')
            temp1.append(date_time_obj)
        else:
            temp1.append(0)

    for i in expire_date:
        if i!='0':
            date_time_obj = datetime.datetime.strptime(i[2:-2],'%Y-%m-%d')
            temp2.append(date_time_obj)
        else:
            temp2.append(0)
    period_time = []
    for i in range(len(temp1)):
        if temp1[i]!=0:
            d0 = temp1[i]
            d1 = temp2[i]
            gap = d1-d0
            days = gap.days
            period_time.append(days)
        else:
            period_time.append(0)
    return period_time

certificate_period = transfer_start_expire_toperiod(date_created_part1,date_expired_part1)
```

```python
import numpy as np
from sklearn.base import TransformerMixin
from sklearn.preprocessing import StandardScaler

class NDStandardScaler(TransformerMixin):
    def __init__(self, **kwargs):
        self._scaler = StandardScaler(copy=True, **kwargs)
        self._orig_shape = None

    def fit(self, X, **kwargs):
        X = np.array(X)
        # Save the original shape to reshape the flattened X later
        # back to its original shape
        if len(X.shape) > 1:
            self._orig_shape = X.shape[1:]
        X = self._flatten(X)
        self._scaler.fit(X, **kwargs)
        return self

    def transform(self, X, **kwargs):
        X = np.array(X)
        X = self._flatten(X)
        X = self._scaler.transform(X, **kwargs)
        X = self._reshape(X)
        return X

    def _flatten(self, X):
        # Reshape X to <= 2 dimensions
        if len(X.shape) > 2:
            n_dims = np.prod(self._orig_shape)
            X = X.reshape(-1, n_dims)
        return X

    def _reshape(self, X):
        # Reshape X back to it's original shape
        if len(X.shape) >= 2:
            X = X.reshape(-1, *self._orig_shape)
        return X
```

```python
class autoencoder_model():
    def __init__(self,X,encoder_size):
        self.encoder_size = encoder_size
        self.input = Input(shape=(X.shape[1], X.shape[2]))
        self.LSTM_encode1 = LSTM(int(X.shape[2]*4), activation='relu', return_sequences=True, 
                  kernel_regularizer=regularizers.l2(0.00))
        self.LSTM_encode2 = LSTM(X.shape[2], activation='relu', return_sequences=False)
        self.Repeat = RepeatVector(X.shape[1])
        self.LSTM_decode3 = LSTM(X.shape[2], activation='relu', return_sequences=True)
        self.LSTM_decode4 = LSTM(int(X.shape[2]*4), activation='relu', return_sequences=True)
        self.decode_timedistribute = TimeDistributed(Dense(X.shape[2]))
        self.encoder = self.lstm_encoder()
        self.decoder = self.lstm_decoder()
    
    def lstm_encoder(self):
        inputs = self.input
        x1 = self.LSTM_encode1(inputs)
        x2 = self.LSTM_encode2(x1)
        latent_dimension = Dense(self.encoder_size)(x2)
        encoder = Model(self.input, latent_dimension, name="lstm_encoder")
        return encoder
    
    def lstm_decoder(self):
        latent_input = Input(shape=(self.encoder_size))
        expend_layer = self.Repeat(latent_input)
        x3 = self.LSTM_decode3(expend_layer)
        x4 = self.LSTM_decode4(x3)
        output = self.decode_timedistribute(x4)
        decoder = Model(latent_input, output, name = 'lstm_decoder')
        return decoder
        
    def lstm_autoencoder(self):
        lstm_aotoencoder = Model(self.input, self.decoder(self.encoder(self.input)),name = "lstm_autoencoder")
        return lstm_aotoencoder
    
    def model_structure(self):
        lstm_aotoencoder = self.lstm_autoencoder()
        self.encoder.summary()
        self.decoder.summary()
        lstm_aotoencoder.summary()
        
    def save_model(self,rootpath):
        auto = self.lstm_autoencoder()
        enco = self.encoder
        deco = self.decoder
        auto_path = rootpath + '/auto'
        enco_path = rootpath + '/enco'
        deco_path = rootpath + '/deco'
        auto.save(auto_path)
        enco.save(enco_path)
        deco.save(deco_path)
        
    def load_model(self,rootpath):
        auto_path = rootpath + '/auto'
        enco_path = rootpath + '/enco'
        deco_path = rootpath + '/deco'
        auto = load_model(auto_path,compile=False)
        enco = load_model(enco_path,compile=False)
        deco = load_model(deco_path,compile=False)
        return auto,enco,deco
        
        
    def train(self, X_train_final,nb_epochs,batch_size):
        auto = self.lstm_autoencoder()
        enco = self.encoder
        deco = self.decoder
        lr_schedule = keras.optimizers.schedules.ExponentialDecay(initial_learning_rate=1e-3,decay_steps=10000,decay_rate=0.9)
        adam = keras.optimizers.Adam(learning_rate=lr_schedule)
        auto.compile(optimizer=adam, loss='mae')
        history = auto.fit(X_train_final, X_train_final, epochs=nb_epochs, batch_size=batch_size,
                    validation_split=0.05,callbacks=[tensorboard_callback]).history
        return enco,deco  
```

```python
import matplotlib.pyplot as plt
import numpy as np
def confidence_visualization(result_df,target_object,threshold):
    score_all = list(result_df[target_object])
    x = range(len(result_df))
    y = score_all
    plt.figure(figsize=(10,10))
    plt.scatter(x, y,s=5)
    theta = result_df['Threshold'][:len(x)]
    plt.plot(x, theta,'r',label='threshould')
    plt.ylim(0, threshold)
    plt.xlabel('object')
    plt.ylabel(target_object)
    plt.legend()
    plt.show()
```
